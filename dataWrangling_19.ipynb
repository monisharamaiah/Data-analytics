```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

1. The data consists of 40 unique records.
2. The cryptocurrency with the symbol "BTC" (Bitcoin) dataset.
3. There is high variability indicating that cryptocurrencies in the dataset have vastly different market caps. 

create scatter plot between name and prize
filter data if prize is <2usd

```python
import matplotlib.pyplot as plt
Name=(coin_df['name'])
prize=(coin_df['priceUsd'])
plt.scatter(Name,prize,color='k',label="name vs prize ",marker='D')
plt.xlabel("Name")
plt.ylabel("Prizes")
plt.grid()
plt.legend()
```

```python
filt=(coin_df[coin_df['priceUsd']<2])
filt.head()
```

```python
import matplotlib.pyplot as plt
Name=(filt['name'])
prize=(filt['priceUsd'])
plt.figure(figsize=(20,5))
plt.scatter(Name,prize,color='k',label="name vs prize ",marker='D')
plt.xlabel("Name")
plt.ylabel("Prizes")
plt.grid()
plt.legend()
```
