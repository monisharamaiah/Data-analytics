# seaborn

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
mydata={'Name':['ram','shyam','bhama','shanti'],
       'Age':[23,45,28,23],
        'salary':[12000,22000,30000,36000],
        'EXP':[2,5,4,1]}
df=pd.DataFrame(mydata)
df.head()
```

histogram

```python
plt.figure(figsize=(2,2))
sns.histplot(df['salary'],kde=True,bins=3)
plt.title('Distribution of salary')
plt.show()
```

1.Negetive skew -->smaller value appeared                                                                                                                     
2.The average salary is observed in range of 25000-30000                                                                                                     
3.average salary can be 28000                                                                                                                            
4.No outlets detected

correlation matrix(heatmap)

```python
ndf=df.select_dtypes(include=['number'])
ndf.head()
```

```python
plt.figure(figsize=(2,2))
sns.heatmap(ndf.corr(),cmap='BuPu',annot=True)
plt.title('Corelation between salary age and exp')
plt.show()
```

1. it shows the Corelation between salary age and exp
2. there is a positive relation between age and exp
3. there is a less relation between age and sal

```python
plt.figure(figsize=(3,3))
sns.boxplot(x=df['Age'])
plt.title('Age dist')
plt.show()
```

1. average age value is 25
2. no abnormal outliner is found

```python
temp=[21,47,39,22,31,33,29,26,27,25,49,46]
plt.figure(figsize=(3,3))
sns.boxplot(x=temp)
plt.title('Temperature distribution')
plt.show()
```

1.Average shows around 30
